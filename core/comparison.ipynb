```python
import pandas as pd

df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)

for col in df1.columns:
    # Check if column exists in both
    if col not in df2.columns:
        print(f"❌ Column '{col}' not found in df2")
        continue

    # Boolean mask where values differ
    mask = (df1[col] != df2[col]) & ~(df1[col].isna() & df2[col].isna())

    if not mask.any():
        print(f"✅ SAME — Column '{col}'")
    else:
        print(f"\n❌ DIFFERENT — Column '{col}'")
        diff = pd.DataFrame({
            'row_index': mask[mask].index,
            'df1_value': df1.loc[mask, col],
            'df2_value': df2.loc[mask, col]
        }).head(5)
        print(diff)
```

```python
import pandas as pd
import os
import glob
import shutil
```

```python
main_folder = r'D:\eSparks\ml_nonquad_model\analysis.noquads.atr.251009.10'
dest_folder = r'D:\eSparks\ml_nonquad_model\Ibrahim\selected_files'

long_pattern = "*_long.renko_trades.all.v3.4l.top_1.csv"
short_pattern = "*_short.renko_trades.all.v3.4l.top_1.csv"

os.makedirs(dest_folder, exist_ok=True)

symbol_dirs = [
    os.path.join(main_folder, d)
    for d in os.listdir(main_folder)
    if os.path.isdir(os.path.join(main_folder, d))
]

for sym_dir in symbol_dirs:
    long_files = glob.glob(os.path.join(sym_dir, long_pattern))
    short_files = glob.glob(os.path.join(sym_dir, short_pattern))

    for f in long_files + short_files:
        try:
            shutil.copy2(f, dest_folder)
            print(f"Copied: {f}")
        except Exception as e:
            print(f"Error copying {f}: {e}")

print(" Copy completed.")
```
